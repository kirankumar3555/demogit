```python
age=np.random.choice([15,20,30,45,12,'10',15,'34',7,'missing'],50)
fico=np.random.choice(['100-150','150-200','200-250','250-300'],50)
city=np.random.choice(['Mumbai','Delhi','Chennai','Kolkata'],50)
ID=np.arange(50)
rating=np.random.choice(['Excellent','Good','Bad','Pathetic'],50)
balance=np.random.choice([10000,20000,30000,4000,np.nan,50000,60000],50)
children=np.random.randint(high=5,low=0,size=(50,))
```

```python
dict1={'ID':ID,
         'fico':fico,
         'city': city,
         'age':age,
         'rating':rating,
         'balance':balance,
         'children':children}
```

```python
df2=pd.DataFrame(dict1)
df2
```

```python
df2['age']=pd.to_numeric(df2['age'],errors='coerce')
```

```python
df2
```

```python
df2['age'].mean()
np.mean(df2['age'])
```

```python
df2[df2['age'].isnull()]
df2.loc[df2['age'].isnull(),'age']=np.mean(df2['age'])
df2
```

```python
df2.loc[df2['balance'].isnull(),'balance']=np.mean(df2['balance'])
df2
```

```python
df2['fico']=pd.Series(fico)
df2
```

```python
fico_list1=list(df2['fico'])
fico_list2=[]
for i in fico_list1:
    i1=i.split('-')
    mean=(float(i1[0])+float(i1[1]))/2
    fico_list2.append(mean)
fico_list2
```
